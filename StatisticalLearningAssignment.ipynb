```python
import numpy as np
import pandas as pd
```

```python
#Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data. (5 points)

LaLiga=pd.read_csv('Laliga.csv',header=0)
LaLiga.replace('-',0,inplace=True)
LaLiga.head(5)
```

```python
#Print all the teams which have started playing between 1930-1980. Use “Debut” column (5 points)

val_gt_1930 = LaLiga['Debut'].values >'1930' 
val_lt_1980 = LaLiga['Debut'].values < '1980'

LaLiga[val_gt_1930 & val_lt_1980][['Team','Debut']].style.hide_index()
```

```python
#Print the list of teams which came Top 5 in terms of points (2.5 points)

LaLiga['Points'] = LaLiga['Points'].astype('int')
LaLiga.sort_values(by='Points',ascending=False,axis=0).head(5)[['Team','Points']].style.hide_index()
```

```python
#Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. 
#Using the same function, find the team which has maximum and minimum goal difference. (5 points)
#Goal_diff_count = GoalsFor - GoalsAgainst

def Goal_diff_count():
    LaLiga['Goal_diff_count'] = LaLiga['GoalsFor'].astype('int') - LaLiga['GoalsAgainst'].astype('int')
    var_max = LaLiga['Goal_diff_count'].max()
    var_min = LaLiga['Goal_diff_count'].min()
    print(f'Maximum Goal difference is', var_max, 'by team - ', LaLiga[LaLiga['Goal_diff_count'] == var_max]['Team'].values)
    print(f'Minimum Goal difference is', var_min, 'by team - ', LaLiga[LaLiga['Goal_diff_count'] == var_min]['Team'].values)
    return LaLiga[['Team','Goal_diff_count']].style.hide_index()
Goal_diff_count()
```

```python
#Create a new column with the name “Winning Percent” and append it to the data set (5 points)
#Percentage of Winning = (GamesWon / GamesPlayed)*100
#If there are any numerical error, replace it with 0%
#Print the top 5 teams which have the highest Winning percentage

LaLiga['Winning Percent'] = round((LaLiga['GamesWon'].astype('int') / LaLiga['GamesPlayed'].astype('int'))*100,2)
LaLiga['Winning Percent'].fillna(0,inplace=True)
LaLiga.sort_values(by='Winning Percent',ascending=False,axis=0).head(5)[['Team','Winning Percent']].style.hide_index()
```

```python
# Group teams based on their “Best position” and print the sum of their points for all positions (7.5 points)

LaLiga.groupby(['BestPosition'])[['Points']].sum()
```
