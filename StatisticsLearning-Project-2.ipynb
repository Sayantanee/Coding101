<span style="font-family: Arial;font-weight:bold;font-size:1.5em;color:#00b3e5;">Step 4: Finding Statistical Evidence

<span style="font-family: Arial;font-size:1.5em;color:#00b3e5;"> a. Do charges of people who smoke differ significantly
from the people who don't?

Divide the smokers and non smokers into two groups

```python
Ins_df_Smokers=Ins_df[Ins_df['smoker']=='yes']['charges'].values
Ins_df_NonSmokers=Ins_df[Ins_df['smoker']=='no']['charges'].values
```

The below mean values for both the groups (Smokers charges :u1 and Non-Smokers Charges :u2) are for our reference

```python
Ins_df_Smokers.mean(),Ins_df_NonSmokers.mean(),Ins_df_Smokers.shape,Ins_df_NonSmokers.shape
```

### State the null and alternative hypothesis:

#### Ho : The mean Charges of Smokers and non Smokers are equal

#### Ha : The mean Charges of Smokers and non Smokers are NOT equal

### Decide Significance Level

Significance level:  α = 0.05

### Calculate t-statistics and p-value

```python
from scipy.stats import ttest_1samp, ttest_ind, mannwhitneyu, levene, shapiro
from statsmodels.stats.power import ttest_power
t_statistic, p_value = ttest_ind(Ins_df_Smokers,Ins_df_NonSmokers)
print(t_statistic, p_value)
```

### Check EDA evidence as well

```python
plt.figure(figsize=(10,5))
sns.stripplot(Ins_df['charges'], Ins_df['smoker'])
```

### Conclusion

Statistical Evidence : p_value<0.05, hence reject Ho. Hence the charges for smoker and non smoker varies significantly.

EDA: The above strip plot also shows significant difference between the charges

<span style="font-family: Arial;font-size:1.5em;color:#00b3e5;"> b. Does bmi of males differ significantly from that of
females?

```python
Ins_df_male=Ins_df[Ins_df['sex']=='male']['bmi'].values
Ins_df_female=Ins_df[Ins_df['sex']=='female']['bmi'].values
```

### State the null and alternative hypothesis:

#### Ho: Average BMI of males and females are same

#### Ha: Average BMI of males and females are NOT same

The above stripplot also depicts the same.

<span style="font-family: Arial;font-size:1.5em;color:#00b3e5;"> c. Is the proportion of smokers significantly different
in different genders?

### State the null and alternative hypothesis:

#### Ho = The proportions are equal
#### Ha = The proportions are not equal

### Decide Significance Level

Significance level:  α = 0.05

```python
female_smokers = Ins_df[Ins_df['sex'] == 'female'].smoker.value_counts()[1]  # number of female smokers
male_smokers = Ins_df[Ins_df['sex'] == 'male'].smoker.value_counts()[1] # number of male smokers
n_females = Ins_df.sex.value_counts()[1] # number of females in the data
n_males = Ins_df.sex.value_counts()[0] #number of males in the data
```

```python
Ins_df.sex.value_counts()
```

```python
print([female_smokers, male_smokers] , [n_females, n_males])
print(f' Proportion of smokers in females, males = {round(115/662,4)*100}%, {round(159/676,4)*100}% respectively')
```

### Test of proportion

```python
from statsmodels.stats.proportion import proportions_ztest

stat,pval = proportions_ztest([female_smokers, male_smokers] , [n_females, n_males])

print(stat,pval)
```

### Conclusion

Since p_value < 0.05, we reject the Ho, which means the proportions of smokers, gender wise, are not equal

<span style="font-family: Arial;font-size:1.5em;color:#00b3e5;"> d. Is the distribution of bmi across women with no
children, one child and two children, the same ?

### Divide the female bmis into 3 groups and find out the mean, std deviation and count of observations

```python
Ins_df_No_children=Ins_df[(Ins_df['children'] == 0) & (Ins_df['sex'] == 'female')]['bmi']
Ins_df_1_child=Ins_df[(Ins_df['children'] == 1) & (Ins_df['sex'] == 'female')]['bmi']
Ins_df_2_children=Ins_df[(Ins_df['children'] == 2) & (Ins_df['sex'] == 'female')]['bmi']
```

```python
print('Count, Mean and standard deviation of mean bmi of Female with no children: %3d, %3.2f and %3.2f' % (len(Ins_df_No_children ), Ins_df_No_children .mean(),np.std(Ins_df_No_children ,ddof =1)))
print('Count, Mean and standard deviation of mean bmi of Female with 1 child: %3d, %3.2f and %3.2f' % (len(Ins_df_1_child), Ins_df_1_child.mean(),np.std(Ins_df_1_child,ddof =1)))
print('Count, Mean and standard deviation of mean bmi of Female with 2 children: %3d, %3.2f and %3.2f' % (len(Ins_df_2_children), Ins_df_2_children.mean(),np.std(Ins_df_2_children,ddof =1)))
```

### Create a single dataframe with group type and respective bmi measures

```python
mean_bmi_df = pd.DataFrame()

df1            = pd.DataFrame({'Female_no_of_children': '0', 'Bmi':Ins_df_No_children})
df2            = pd.DataFrame({'Female_no_of_children': '1', 'Bmi':Ins_df_1_child})
df3            = pd.DataFrame({'Female_no_of_children': '2', 'Bmi':Ins_df_2_children})

mean_bmi_df = mean_bmi_df.append(df1) 
mean_bmi_df = mean_bmi_df.append(df2) 
mean_bmi_df = mean_bmi_df.append(df3) 
```

```python
mean_bmi_df.head()
```

### State the null and alternative hypothesis:

* $H_0$: $\mu1$ = $\mu2$ = $\mu3$
* $H_A$: At least one $\mu$ differs 

Here $\mu1$, $\mu2$ and $\mu3$ are the mean bmi for female with no children, female with 1 child and female with 2 children respectively.

### Decide Significance Level

Significance level:  α = 0.05

### Calculate p value using ANOVA table

```python
import statsmodels.api         as     sm
from   statsmodels.formula.api import ols
 
mod = ols('Bmi ~ Female_no_of_children', data = mean_bmi_df).fit()
aov_table = sm.stats.anova_lm(mod, typ=2)
print(aov_table)
```

### Check with the help of boxplot as well

```python
sns.boxplot(x = "Female_no_of_children", y = "Bmi", data = mean_bmi_df)
plt.title('BMI of different Female groups')
plt.show()
```

### Conclusion

Statistical evidence: SInce p-value is > 0.05, our significance level, thus we accept the Ho (Null Hypothesis)

Also, the boxplot depicts that the  distributions of ‘bmi’ are nearly same across women with 0, 1 or 2 children.
